# Hand-Made Central Limit Theorem

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
%matplotlib inline
```

Central Limit Theorem says that if we have a distribution $F(X)$ with expected value $\mu$ and dispersion $\sigma^2$,  and we select $n$  i.i.d. random variables $X_1, X_2, ..., X_n$, then their mean $\bar{X}$ is distributied normally with dispersion  $\sigma^2/n$ and mean $\mu$:

$$\bar{X}_n \approx N(\mathbb{E}\mu, \sigma^2 / n)$$

Let us check if it really so for the case when $F(X)$ is a uniform distribution:

Define uniform distribution from a to b:

```python
a = 1
b = 5

uniform_rv = sts.uniform(a, b-a)
uniform_rv.rvs(10)
```

**Task 1**. Plot the probability density function of the uniform distribution:

```python
#your code here
x = np.linspace(0,6,100)#creates linespace for graph range, 0 to 6 so that we can see the areas where prob is zero.
pdf = uniform_rv.pdf(x)#finds pdf from uniform distribution
plt.plot(x,pdf) #plots pdf and x axis
plt.title("Probability Density function for Uniform Distribution")#sets title
plt.ylabel("Frequency")#sets y label

#you should obtain something like this:
```

We are going to create *samples_no = 1000* sets of the size *n = 50* and calculate mean $\bar{X}$ of each sample and store the results in the vector *sample_means*:

```python
samples_no = 1000
```

```python

n = 50
sample_means = np.empty(samples_no)
```

**Task 2**: Create a cycle for *i* from 0 to *samples_no*, in which you create a sample of the size *n*, calculate the mean of the sample and store its value in the vector *sample_means*

```python
#your code here
for i in range(0,samples_no):

    sample = np.random.rand(50,1) #creates a sample of 50 random numbers
    storage = sample.mean() #finds the mean of the sample
    sample_means[i] = storage #stores the mean of the sample into a array
```

```python
fig, ax = plt.subplots() #sets up plot
bins= ax.hist(sample_means) #plots a histogram of plot
ax.set_ylabel("fraction of means") #axis labels
ax.set_xlabel("x")
```

You are going to compare the obtained histogram from Task 3 with the normal distribution

**Task 4:** Create a normal distribution such, that it will be a distribution of the data you store in the vector *sample_means* and plot its density distribution function. Hint: to do that, you should know the **theoretical** expected value (i.e. mean) and dispersion (i.e. variance) of your initial uniform distribution. Look it up in Wikipedia.

```python
# your code here:
#we need to find the normal distribution modelling our data, so we need the mean and standard deviation of the set

bin_x = np.linspace(bins[1].min(),bins[1].max()) #creates a x axis using the bins from the previous task


mean_height = sample_means.mean() #finds the mean and standard deviation of the random samples
std_height = np.std(sample_means)

y = (1/(std_height*np.sqrt(2*math.pi)))*np.exp((-1/2)*((bin_x-mean_height)/std_height)**2) #fits the normal distribution using the mean and standard deviation

fig2, ax2 = plt.subplots() #sets up plot again
ax2.plot(bin_x,y) #plots normal distribution
```

**Task 5** Plot the histogram from Task 3 and the distribution density function from Task 4 *in the same Graph*. Do they look similar?

```python
fig3, ax3 = plt.subplots()
ax3.plot(bin_x,y) #plots normal distribution
bins= ax3.hist(sample_means,density = "true", bins = 30) #plots histogram
#the two look very similar, variation from the limiting distribution can be attributed to the random function used.
```
